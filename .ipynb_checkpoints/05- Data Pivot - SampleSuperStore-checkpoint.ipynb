# Sample Super Store Part  - Data Pivoting
## This is a sample notebook from the Sample Super Store Dataset

This notebook is intended to give a brief overview on how to explore dataset, which in this case is the Ssample Super Store dataset, which I believe to be a fictional ecommerce company. With that being said, let's try to explore certain data angle with the Panda's library for Python using the Jupyter Notebook.

### Importing the required libraries into the notebook.

```python
import pandas as pd
import numpy as np
```

### Let's try to read the CSV record 
The following code would imply these instructions
- df_orders = is the name of the variable, that will be using throughout the example of this tutorial.
- pd = stands for Panda
- .read_csv = is a method within to read the CSV file.

```python
df_cleaning = pd.read_csv('df_cleaning.csv')
```

```python
# Set the total rows and columns default numbers.
pd.set_option('display.max_columns', 23)
pd.set_option('display.max_rows', 5)
```

```python
df_cleaning.head(3)
```

```python
df_cleaning.pivot_table(df_cleaning,index=["OrderID"])
```

```python
df_cleaning.pivot_table(df_cleaning,index=["OrderID","CustomerName","ProductName", "Sales", "Quantity", "Discount", "Profit", "Revenue", "NetPrice"])
```

```python
df_cleaning.pivot_table(df_cleaning,index=["OrderID","CustomerName","ProductName"])
```

```python
df_cleaning.pivot_table(index=["OrderID","CustomerName"],values=["Sales"])
```

```python
df_cleaning.pivot_table(index=["OrderID","CustomerName"], values=["Sales"], columns = ['Segment', "Category"], aggfunc = 'sum')
```

```python
df_cleaning.pivot_table(index=["OrderID","CustomerName"],values=["NetPrice"],aggfunc=np.sum)
```

```python
df_cleaning.pivot_table(index=["CustomerName","OrderID"],values=["NetPrice"],aggfunc=[np.sum,len])
```

```python
df_cleaning.pivot_table(index=["OrderID","CustomerName"],values=["NetPrice"],aggfunc=[np.mean,len])
```

```python
df_cleaning.pivot_table(index=["CustomerName","OrderID"],values=["Sales"],columns=["Segment", "Category", "SubCategory"],aggfunc=[np.sum])
```

```python
df_cleaning.pivot_table(index=["CustomerName","OrderID"],values=["Sales"],columns=["Segment", "Category", "SubCategory"],aggfunc=[np.sum], fill_value=0)
```

```python
df_cleaning.pivot_table(index=["CustomerName","OrderID"],values=["Sales", "Quantity", "NetPrice"],
                        columns=["Segment", "Category", "SubCategory"],
                        aggfunc=[np.sum], 
                        fill_value=0)
```

```python
df_cleaning.pivot_table(index=["CustomerName","OrderID","ProductName"],
                        values=["Sales", "Quantity", "NetPrice"],
                        columns=["Segment", "Category"],
                        aggfunc=[np.sum], 
                        fill_value=0)
```

```python
df_cleaning.pivot_table(index=["CustomerName","OrderID","ProductName"],
                        values=["Sales", "Quantity", "NetPrice"],
                        columns=["Segment", "Category"],
                        aggfunc=[np.sum], 
                        fill_value=0, margins=True)
```

```python
df_cleaning.pivot_table(index=["CustomerName"],
                        values=["Sales", "Quantity", "NetPrice"],
                        columns=["Segment", "Category"],
                        aggfunc=[np.sum], 
                        fill_value=0, margins=True)
```

```python
df_cleaning.pivot_table(index=["ShipMode"],
                        values=["Sales", "Quantity", "NetPrice"],
                        columns=["Segment", "Category"],
                        aggfunc=[np.sum], 
                        fill_value=0, margins=True)
```

```python
table = df_cleaning.pivot_table(index=["ProductName"],
                                columns=["Category", "SubCategory"],
                                values=["Quantity","NetPrice"],
                                aggfunc={"Quantity":len, "NetPrice":[np.sum,np.mean]},
                                fill_value=0)
table
```

```python
table.query('ProductName == ["Bush Somerset Collection Bookcase"]')
```

```python
table.query('ProductName == ["Bush Somerset Collection Bookcase", "Eldon Fold Roll Cart System"]')
```
