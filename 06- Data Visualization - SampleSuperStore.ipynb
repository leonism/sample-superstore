# Seaborn Visual Library

We'll try to illustrate some charts based on couple of columns throughout the dataset with the Seaborn library. Let's try something simple with only a single X axis values againts its own value counts.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Let's try to visualize the 'Segment' column from the dataset.

# Calculate the counts for each segment
segment_counts = df_cleaning['Segment'].value_counts()

# Create a bar plot with segment counts
fig, ax = plt.subplots(figsize=(10, 6))
sns.countplot(x='Segment', data=df_cleaning, palette='rainbow', ax=ax)

# Add annotations for each bar with its count
for p in ax.patches:
    height = p.get_height()
    ax.text(p.get_x() + p.get_width() / 2, height + 3, f'{height:.0f}', ha='center', va='bottom')

# Set title and labels for better understanding
ax.set_title('Market Segmentation')
ax.set_xlabel('Segment')
ax.set_ylabel('Transactions')

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set the style for the plot
sns.set_style('whitegrid')

# Calculate the count for each segment
segment_counts = df_cleaning['Segment'].value_counts()

# Create a bar plot with segment counts
sns.barplot(x=segment_counts.index, y=segment_counts.values)

# Add a title and labels for better understanding
plt.title('Count of Segments')
plt.xlabel('Segment')
plt.ylabel('Count')

# Show the plot
plt.show()
```

```python
# Let's try to visualize the 'Category' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='Category',data=df_cleaning, palette='rainbow', legend=False)
```

```python
# Let's try to visualize the 'SubCategory' column from the dataset.

sns.set_style('whitegrid')
sns.countplot(x='SubCategory',data=df_cleaning, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='Region',data=df_cleaning, palette='rainbow')
```

```python
sns.set_style('whitegrid')
sns.countplot(x='ShipMode',data=df_cleaning, palette='rainbow')
```

```python
# Let's try to visualize the 'Region' column from the Shipmode column dataset standpoints.

region_hist = sns.FacetGrid(df_cleaning, col='ShipMode', palette='rainbow')
region_hist.map(plt.hist, 'Region')
region_hist.set_ylabels('Number')
```

```python
# Let's try to visualize the 'Category' column from the Shipmode column dataset standpoints.

category_hist = sns.FacetGrid(df_cleaning, col='ShipMode', palette='rainbow')
category_hist.map(plt.hist, 'Category')
category_hist.set_ylabels('Number')
```

```python
# Let's try to visualize the 'SubCategory' column from the Shipmode column dataset standpoints.

subcategory_hist = sns.FacetGrid(df_cleaning, col='Segment', height=10.5, aspect=4.6)
subcategory_hist.map(plt.hist, 'SubCategory')
subcategory_hist.set_ylabels('Number')
```

```python
# Let's try to visualize the 'Category' column from the Shipmode column dataset standpoints.

pclass_age_grid = sns.FacetGrid(df_cleaning, col='Category', row='SubCategory', height=2.2, aspect=1.6)
pclass_age_grid.map(plt.hist, 'Segment')
pclass_age_grid.set_ylabels('Number')
pclass_age_grid.add_legend()
```

```python
grid = sns.FacetGrid(df_cleaning, row='Category', col='SubCategory', height=2.2, aspect=1.6)
grid.map(sns.barplot, 'NetPrice', 'Segment', alpha=.5, ci=None)
grid.add_legend()
```
